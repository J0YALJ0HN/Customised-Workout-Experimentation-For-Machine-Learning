```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'bodyPerformance_with_difficulty_v2.csv'
df = pd.read_csv(file_path)

# Convert gender to numerical values using label encoding
df['gender'] = df['gender'].map({'M': 0, 'F': 1})

# Select relevant columns
selected_columns = ['age', 'gender', 'height_cm', 'weight_kg', 'bodyFat_', 'gripForce',
                     'sitAndBendForward_cm', 'situpsCounts', 'broadJump_cm', 'workout_difficulty']

# Create a correlation matrix
correlation_matrix = df[selected_columns].corr()

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

# Show the plot
plt.title('Correlation Matrix of Features with Workout Difficulty')
plt.show()
```

This code uses the seaborn.heatmap function to create a visual representation of the correlation matrix. The values in the matrix range from -1 to 1, where 1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no correlation. The color intensity in the heatmap represents the strength and direction of the correlation.
